# 구내식당 중식 인원 머신러닝

+ 1. Use train data and check performance of each algorithm
+ 2. Apply some best algorithm to test data

```python
ln_train = pd.read_csv("../byun/data/lunch_df_encoding.csv", encoding="utf-8")
ln_train
```

```python
ln_train.info()
ln_train.shape
```

```python
ln_train.isnull().sum() #칼럼별 결측치 수
```

```python
ln_train = ln_train.dropna()
ln_train
```

## 선형회귀(일반) - train: 78.40%, test: 81.09%

```python
X= ln_train.drop(columns=["datetime", "year", "month", "date", "lunch_number"], axis=1)

y = ln_train["lunch_number"]

y = y.to_numpy()
y
X.shape
```

```python
lr = LinearRegression().fit(X, y)
print("훈련 점수 : ", lr.score(X, y))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
lr = LinearRegression().fit(X_train, y_train)
```

```python
print("훈련 점수 : ", lr.score(X_train, y_train))
print("테스트 점수 : ", lr.score(X_test, y_test))
```

## 릿지회귀모델 - train: 78.42%, test: 80.96%

```python
from sklearn.linear_model import Ridge

ridge = Ridge().fit(X_train, y_train) # Ridge(alpha=1.0) 이 디폴트

print("훈련 점수 : ", ridge.score(X_train, y_train))
print("테스트 점수 : ", ridge.score(X_test, y_test))

ridge10 = Ridge(alpha=10).fit(X_train, y_train) # Ridge(alpha=1.0) 이 디폴트

print("훈련 점수 : ", ridge10.score(X_train, y_train))
print("테스트 점수 : ", ridge10.score(X_test, y_test))


ridge01 = Ridge(alpha=0.000000001).fit(X_train, y_train) # Ridge(alpha=1.0) 이 디폴트

print("훈련 점수 : ", ridge01.score(X_train, y_train))
print("테스트 점수 : ", ridge01.score(X_test, y_test))

# Ridge: lower alpha makes greater score
```

## 라쏘회귀모델 - train: 78.41%, test: 80.95%

```python
from sklearn.linear_model import Lasso

lasso = Lasso().fit(X_train, y_train) #Lasso(alpha=1) 디폴트

print("훈련 점수 : ", lasso.score(X_train, y_train))
print("테스트 점수 : ", lasso.score(X_test, y_test)) 

print("사용한 특성의 수 : ", np.sum(lasso.coef_ != 0)) 
lasso001 = Lasso(alpha=0.01).fit(X_train, y_train)

print("훈련 점수 : ", lasso001.score(X_train, y_train))
print("테스트 점수 : ", lasso001.score(X_test, y_test))

print("사용한 특성의 수 : ", np.sum(lasso001.coef_ != 0))

lasso00001 = Lasso(alpha=0.000000001).fit(X_train, y_train)

print("훈련 점수 : ", lasso00001.score(X_train, y_train))
print("테스트 점수 : ", lasso00001.score(X_test, y_test)) 

print("사용한 특성의 수 : ", np.sum(lasso00001.coef_ != 0))

# Lasso: lower alpha makes greater score
```

## SVR model - train: 77.53%, test: 80.27%

- Normal
- Kernel("linear" is best)
- StandardScaler // MinMaxScaler

```python
model = SVR()  # default is rbf
model.fit(X_train, y_train)

print("학습 결과 : ", model.score(X_train, y_train))
print("테스트 결과 : ", model.score(X_test, y_test))

# Too bad
```

```python
linear_svr = SVR(kernel="linear")
linear_svr.fit(X_train, y_train)

print("학습 결과 : ", linear_svr.score(X_train, y_train))
print("테스트 결과 : ", linear_svr.score(X_test, y_test))

print('===================================================')

linear_svr = SVR(kernel="linear", C=10, gamma='scale')
linear_svr.fit(X_train, y_train)

print("학습 결과 : ", linear_svr.score(X_train, y_train))
print("테스트 결과 : ", linear_svr.score(X_test, y_test))

# Greater than previous model but it is even lower than lasso and ridge

# print('===================================================')
# polynomial_svr = SVR(kernel="poly")
# polynomial_svr.fit(X_train, y_train)

# print("학습 결과 : ", polynomial_svr.score(X_train, y_train))
# print("테스트 결과 : ", polynomial_svr.score(X_test, y_test))
```

```python
### 정규화 해보자

scaler = MinMaxScaler()
X_train_mms = scaler.fit_transform(X_train)
X_test_mms = scaler.transform(X_test)

model = SVR(kernel='linear')
model.fit(X_train_mms, y_train)

print("학습 결과 : ", model.score(X_train_mms, y_train))
print("테스트 결과 : ", model.score(X_test_mms, y_test))

## 표준화든 정규화든 크게 관계없이 성능이 비슷한 수준으로 좋아졌다
```

```python
### 표준화 해보자

scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)

model = SVR(kernel='linear')
model.fit(X_train_std, y_train)

print("학습 결과 : ", model.score(X_train_std, y_train))
print("테스트 결과 : ", model.score(X_test_std, y_test))

## 표준화만 했는데도 성능이 좋아짐
```

```python
model = SVR(kernel='linear', C=10)
model.fit(X_train_std, y_train)

print("학습 결과 : ", model.score(X_train_std, y_train))
print("테스트 결과 : ", model.score(X_test_std, y_test))
print('===================================================')
model = SVR(kernel='linear', C=10, gamma='auto')
model.fit(X_train_std, y_train)

print("학습 결과 : ", model.score(X_train_std, y_train))
print("테스트 결과 : ", model.score(X_test_std, y_test))
print('===================================================')
# gamma=scale & gamma=auto are same score
```

```python
from sklearn.pipeline import Pipeline

pipe = Pipeline([("scaler", StandardScaler()), ("model", SVR(kernel="linear"))])
params = [
    {
        "model__gamma":["scale", "auto"],
        "model__C":[10.0, 1.0, 0.1, 0.01, 0.001],
        "model__epsilon":[10.0, 1.0, 0.1, 0.01, 0.001]
    }
]

gs = GridSearchCV(
    estimator=pipe,
    param_grid=params,
    n_jobs=multiprocessing.cpu_count(),
    cv=5,
    verbose=True
)
gs.fit(X_train_std, y_train)
```

```python
# gs로 best 값 찾기

print(gs.best_params_)
print(gs.best_estimator_)
print(gs.best_score_)
```

```python
# best 값 적용
model = SVR(kernel="linear", C=10.0, gamma="scale", epsilon=10.0)
model.fit(X_train_std, y_train)

print("학습 결과 : ", model.score(X_train_std, y_train))
print("테스트 결과 : ", model.score(X_test_std, y_test))
```

### XGBoostRegressor - train: 99.76%, test: 83.62%, testscore_mean: 76.23%

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=0)

xgbr = XGBRegressor(n_estimators=300, learning_rate=0.08, max_depth=6, objective="reg:squarederror")
xgbr.fit(X_train, y_train)

print("훈련 정확도 : ", xgbr.score(X_train, y_train))
print("테스트 정확도 : ", xgbr.score(X_test, y_test))

# Overfitting, but test accuracy is pretty good
```

```python
cross_val = cross_validate(estimator=xgbr, X=X, y=y, cv=15)

print("테스트 정확도 평균 : ", cross_val["test_score"].mean())
```

### LightGBM Regressor - train: 96.09%, test: 82.58%, testscore_mean: 75.84%

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.25, random_state=0)

lgbr = LGBMRegressor(n_estimator=100)
lgbr.fit(X_train, y_train, early_stopping_rounds=50, 
         eval_metric="logloss", eval_set=[(X_test, y_test)], verbose=True)

print("훈련 정확도 : ", lgbr.score(X_train, y_train))
print("테스트 정확도 : ", lgbr.score(X_test, y_test))

cross_val = cross_validate(estimator=lgbr, X=X, y=y, cv=15)
print("테스트 정확도 평균 : ", cross_val["test_score"].mean())
```

## 선형회귀(일반) - train: 52.53%, test: 48.01%

```python
X = dn_train.drop(columns=["datetime", "dinner_number"], axis=1)

y = dn_train["dinner_number"]

y = y.to_numpy()


lr = LinearRegression().fit(X, y)
print("훈련 점수 : ", lr.score(X, y))
print("========================================================")
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
lr = LinearRegression().fit(X_train, y_train)

print("훈련 점수 : ", lr.score(X_train, y_train))
print("테스트 점수 : ", lr.score(X_test, y_test))
```

## 릿지회귀모델 - train: 53.11%, test: 50.38%

```python

ridge = Ridge().fit(X_train, y_train) 

print("훈련 점수 : ", ridge.score(X_train, y_train))
print("테스트 점수 : ", ridge.score(X_test, y_test))
print("========================================================")

ridge10 = Ridge(alpha=100).fit(X_train, y_train) 

print("훈련 점수 : ", ridge10.score(X_train, y_train))
print("테스트 점수 : ", ridge10.score(X_test, y_test))

print("========================================================")
ridge01 = Ridge(alpha=0.000000001).fit(X_train, y_train) 

print("훈련 점수 : ", ridge01.score(X_train, y_train))
print("테스트 점수 : ", ridge01.score(X_test, y_test))
```

## 라쏘회귀모델 - train: 53.10%, test: 50.47%

```python

lasso = Lasso().fit(X_train, y_train) #Lasso(alpha=1) 디폴트

print("훈련 점수 : ", lasso.score(X_train, y_train))
print("테스트 점수 : ", lasso.score(X_test, y_test)) 

print("사용한 특성의 수 : ", np.sum(lasso.coef_ != 0)) 
print("========================================================")
lasso001 = Lasso(alpha=0.01).fit(X_train, y_train)

print("훈련 점수 : ", lasso001.score(X_train, y_train))
print("테스트 점수 : ", lasso001.score(X_test, y_test))

print("사용한 특성의 수 : ", np.sum(lasso001.coef_ != 0))
print("========================================================")
lasso00001 = Lasso(alpha=0.000000001).fit(X_train, y_train)

print("훈련 점수 : ", lasso00001.score(X_train, y_train))
print("테스트 점수 : ", lasso00001.score(X_test, y_test)) 

print("사용한 특성의 수 : ", np.sum(lasso00001.coef_ != 0))

# Lasso: lower alpha makes greater score
```

## SVR model - train: 47.37%, test: 48.92%

- Normal
- Kernel("linear" is best)
- StandardScaler // MinMaxScaler

```python
model = SVR()  # default is rbf
model.fit(X_train, y_train)

print("rbf 학습 결과 : ", model.score(X_train, y_train))
print("rbf 테스트 결과 : ", model.score(X_test, y_test))
print('===================================================')
# Too bad

linear_svr = SVR(kernel="linear")
linear_svr.fit(X_train, y_train)

print("linear 학습 결과 : ", linear_svr.score(X_train, y_train))
print("linear 테스트 결과 : ", linear_svr.score(X_test, y_test))

print('===================================================')

linear_svr = SVR(kernel="linear", C=10, gamma='scale')
linear_svr.fit(X_train, y_train)

print("C=10 학습 결과 : ", linear_svr.score(X_train, y_train))
print("C=10 테스트 결과 : ", linear_svr.score(X_test, y_test))

print('===================================================')
### 정규화 해보자

scaler = MinMaxScaler()
X_train_mms = scaler.fit_transform(X_train)
X_test_mms = scaler.transform(X_test)

model = SVR(kernel='linear')
model.fit(X_train_mms, y_train)

print("minmax 학습 결과 : ", model.score(X_train_mms, y_train))
print("minmax 테스트 결과 : ", model.score(X_test_mms, y_test))

print('===================================================')
### 표준화 해보자

scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)

model = SVR(kernel='linear')
model.fit(X_train_std, y_train)

print("std 학습 결과 : ", model.score(X_train_std, y_train))
print("std 테스트 결과 : ", model.score(X_test_std, y_test))
print('===================================================')
## 표준화만 했는데도 성능이 좋아짐

model = SVR(kernel='linear', C=10)
model.fit(X_train_std, y_train)

print("std 학습 결과 : ", model.score(X_train_std, y_train))
print("std 테스트 결과 : ", model.score(X_test_std, y_test))
print('===================================================')
model = SVR(kernel='linear', C=10, gamma='auto')
model.fit(X_train_std, y_train)

print("std 학습 결과 : ", model.score(X_train_std, y_train))
print("std 테스트 결과 : ", model.score(X_test_std, y_test))
print('===================================================')
# gamma=scale & gamma=auto are same score
```

```python
pipe = Pipeline([("scaler", StandardScaler()), ("model", SVR(kernel="linear"))])
params = [
    {
        "model__gamma":["scale", "auto"],
        "model__C":[10.0, 1.0, 0.1, 0.01, 0.001],
        "model__epsilon":[10.0, 1.0, 0.1, 0.01, 0.001]
    }
]

gs = GridSearchCV(
    estimator=pipe,
    param_grid=params,
    n_jobs=multiprocessing.cpu_count(),
    cv=5,
    verbose=True
)
gs.fit(X_train_std, y_train)

# gs로 best 값 찾기

print(gs.best_params_)
print(gs.best_estimator_)
print(gs.best_score_)
```

```python
# best 값 적용 (linear kernel, C=10, gamma=scale, epsilon=10)
model = SVR(kernel="linear", C=10.0, gamma="scale", epsilon=10.0)
model.fit(X_train_std, y_train)

print("학습 결과 : ", model.score(X_train_std, y_train))
print("테스트 결과 : ", model.score(X_test_std, y_test))
```
