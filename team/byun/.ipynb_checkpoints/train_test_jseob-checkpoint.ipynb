```python
test = pd.read_csv("C:/Users/tagos71/Desktop/Team3/data/test.csv", encoding="utf-8")
test.head(20)
```

```python
test['일자'] = pd.to_datetime(test['일자'])
test.info()
```

```python
test['현본사소속재택근무자수'] = test['현본사소속재택근무자수'].astype(np.int64)
test.head()
```

```python
test['Month'] = test['일자'].dt.strftime('%m')
test['Date'] = test['일자'].dt.strftime('%d')
test['실질정원수'] = test['본사정원수'] - (test['본사휴가자수'] + test['본사출장자수'] + test['현본사소속재택근무자수'])
test['Month'] = test['Month'].astype(np.int64)
test['Date'] = test['Date'].astype(np.int64)

test = test.loc[:, ['일자','Month', 'Date', '요일', '본사정원수', '실질정원수', '본사휴가자수', '본사출장자수', '본사시간외근무명령서승인건수', '현본사소속재택근무자수', '중식메뉴', '석식메뉴']]

test.info()
test.head()
```

```python
### 계절추가
season = []

for index in range(len(test)):
    if test['Month'][index] in [3,4,5]:
        season.append('봄')
    elif test['Month'][index] in [6,7,8]:
        season.append('여름')
    elif test['Month'][index] in [9,10,11]:
        season.append('가을')
    elif test['Month'][index] in [12,1,2]:
        season.append('겨울')

len(season)
```

```python
test['Season'] = season
test
```

```python
New_lunch = []
New_dinner = []

for index in range(len(test)):
    if 'New' in test['중식메뉴'][index]:
        New_lunch.append("Y")
    else:
        New_lunch.append("N")

for index in range(len(test)):
    if 'New' in test['석식메뉴'][index]:
        New_dinner.append("Y")
    else:
        New_dinner.append("N")

print(len(New_lunch))
print(len(New_dinner))
```

```python
test['New_lunch'] = New_lunch
test['New_dinner'] = New_dinner

test
```

```python
# 일별 점심메뉴를 작은 리스트로 갖고 있는 큰 리스트 (lunch) 만들기 
lunch = []

for index in range(len(test)):
    tmp = test.loc[index,'중식메뉴'].split(' ') # 스페이스로 구분
    tmp = ' '.join(tmp).split()    # 빈칸 제거

    # ()안에 있는 내용 제거
    for menu in tmp:
        if '(' in menu:
            tmp.remove(menu)
    lunch.append(tmp) 

for index in range(len(test)): # 밥은 큰 차이가 없을것같으므로 쌀밥은 다 밥으로 통일
    if '쌀밥' in lunch[index][0]:
        lunch[index][0] = '밥'

for i, v in enumerate(lunch, start=1):
    print(i, v)
```

```python
rice=[]
soup=[]
main=[]

for i in range(len(test)):
    if lunch[i][0]:
        rice.append(lunch[i][0])
    if lunch[i][1]:
        soup.append(lunch[i][1])
    if lunch[i][2]:
        main.append(lunch[i][2])

test['lunch_rice'] = rice
test['lunch_soup'] = soup
test['lunch_main'] = main
test
```

```python
dinner = []

for index in range(len(test)):
    tmp = test.loc[index,'석식메뉴'].split(' ')
    tmp = ' '.join(tmp).split()

    for menu_d in tmp:
        if '(' in menu_d:
            tmp.remove(menu_d)
    dinner.append(tmp)

print(len(test))
print(len(dinner))
print(len(lunch))
```

```python
dinner_rice=[]
dinner_soup=[]
dinner_main=[]


for word in dinner:
    
    
    if len(word) == 0:
        dinner_rice.append('None') #비어있으면 그 날짜자체가 없어질수있으므로 'None' 값으로 대체
        dinner_soup.append('None')
        dinner_main.append('None')
    elif '*' in word:
        dinner_rice.append('None') #비어있으면 그 날짜자체가 없어질수있으므로 'None' 값으로 대체
        dinner_soup.append('None')
        dinner_main.append('None')
    elif '가정의날' in word:
        dinner_rice.append('None') #비어있으면 그 날짜자체가 없어질수있으므로 'None' 값으로 대체
        dinner_soup.append('None')
        dinner_main.append('None')
    elif '가정의달' in word:
        dinner_rice.append('None') #비어있으면 그 날짜자체가 없어질수있으므로 'None' 값으로 대체
        dinner_soup.append('None')
        dinner_main.append('None')
    elif '자기계발의날' in word:
        dinner_rice.append('None') #비어있으면 그 날짜자체가 없어질수있으므로 'None' 값으로 대체
        dinner_soup.append('None')
        dinner_main.append('None')
    elif '*자기계발의날*' in word:
        dinner_rice.append('None') #비어있으면 그 날짜자체가 없어질수있으므로 'None' 값으로 대체
        dinner_soup.append('None')
        dinner_main.append('None')
    elif '자기개발의날' in word:
        dinner_rice.append('None') #비어있으면 그 날짜자체가 없어질수있으므로 'None' 값으로 대체
        dinner_soup.append('None')
        dinner_main.append('None')

    else:
        dinner_rice.append(word[0])
        dinner_soup.append(word[1])
        dinner_main.append(word[2])
```

```python
test['dinner_rice'] = dinner_rice
test['dinner_soup'] = dinner_soup
test['dinner_main'] = dinner_main

print(len(dinner))
print(len(dinner_rice))
print(len(dinner_soup))
print(len(dinner_main))
dinner_rice[0]

print(test['dinner_rice'].unique())
```
